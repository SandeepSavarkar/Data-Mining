```python
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
data=pd.read_csv('store_data.csv',header=None);
data.head();
data.shape
```

```python
records = [];
for i in range(0,10):  
    records.append([str(data.values[i,j]) for j in range(0, 7)]);
association_rules = apriori(records, min_support=0.005, min_confidence=0.3, min_lift=2.5,min_length=2);
association_results = list(association_rules);
```

```python
for item in association_results:
    pair = item[0] 
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])
    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    
    print("=====================================")
print(len(association_results)); 
print(type(association_rules));
```
